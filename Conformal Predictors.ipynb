```python
import numpy as np
x = [(0,3),(2,2),(3,3),(-1,1),(-1,-1),(0,1)]
y = [+1,+1,+1,-1,-1,-1]
test = (0,0)

def CP_Diff_Class(X,Y,test):
    if len(X) == len(Y): #check if Samples and labels have same length
        labels = list(set(Y)) #Find out the distinct Labels to assign
        print("Conformity Scores using Nearest Distances to Different Classes\n")
        
        for  k in range(len(labels)): #Loop to iterate through labels i.e consider +1 or -1
                x =  X.copy() 
                y =  Y.copy()
                conformity_scores = [] # to store nearest distance to different class
                distance_array = [] # to store all distances
                
                x.append(test) # Create an Augmented Training Set
                y.append(labels[k]) #Create Augmented set using a label to consider
                print("Considering Label of "+str(test)+" as: ",labels[k])
                
                for i in range(len(x)): #loop to find distances from different classes
                    a1 = x[i][0]
                    b1 = x[i][1]
                    
                    for j in range(len(x)):
                        if j!=i:  # check its not the same sample
                            if y[i] != y[j]: #check if it has a different label
                                a2 = x[j][0]
                                b2 = x[j][1]
                                t1 = (a1-a2)**2
                                t2 = (b1-b2)**2
                                d = np.sqrt(t1+t2) # calculations to find Euclidean Distance
                                distance_array.append(d) #append each distance to different class in this array
                    minimum_index = distance_array.index(min(distance_array)) #Choose the nearest distance index
                    conformity_scores.append(distance_array[minimum_index]) #append nearest Distance to array
                    distance_array.clear() #reset distance array to store next set of distances
                
                test_sample_score = conformity_scores[len(x)-1]
                conformity_scores.sort(reverse=True) #Sort in Descending Order to valculate rank Pessimistically
                for rank in range(len(x)): # Finding rank pessimistically
                    if conformity_scores[rank]==test_sample_score:
                        break
                        
                rank = len(x)-rank #Logic for Pessimistic Approach to find Rank
                print("Rank of Test Sample "+str(test)+" is: "+str(rank))
                
                p_value = (rank)/len(x) #formula for p_value
                print("P_value: "+str(rank)+"/"+str(len(x))+": ",p_value)
                p_value = p_value*100
                p_value = round(p_value,2)
                print("Percentage: "+str(p_value)+" %")
                print("\n")
                
                #Reset all the arrays
                x.clear()
                y.clear()
                conformity_scores.clear()
                distance_array.clear()
        
    else:
        print("Error:Size of Samples and Labels mismatched")
    
```

```python
CP_Diff_Class(x,y,test)
```

```python
import numpy as np
x = [(0,3),(2,2),(3,3),(-1,1),(-1,-1),(0,1)]
y = [+1,+1,+1,-1,-1,-1]
test = (0,0)

def CP_DiffBySame(X,Y,test):
    if len(X) == len(Y): #check if Samples and labels have same length
        labels = list(set(Y)) #Find out the distinct Labels to assign
        print("Conformity Scores using Nearest Distances to Diff/Same Classes\n")
        
        for  k in range(len(labels)): #Loop to iterate through labels i.e consider +1 or -1
                x =  X.copy()
                y =  Y.copy()
                diff_class = [] #to store distances to different classes
                same_class = [] #to store distances to same classes
                distance_array1 = []
                distance_array2 = []
                conformity_scores = [] #to store diff/same class calculations
                
                x.append(test) # Create an Augmented Training Set
                y.append(labels[k]) #Create Augmented set using a label to consider
                print("Considering Label of "+str(test)+" as: ",labels[k])
                
                for i in range(len(x)): #loop to go through all samples
                    a1 = x[i][0]
                    b1 = x[i][1]
                    for j in range(len(x)):
                        if j!=i:
                            a2 = x[j][0]
                            b2 = x[j][1]
                            t1 = (a1-a2)**2
                            t2 = (b1-b2)**2
                            d = np.sqrt(t1+t2) # formula to find Euclidean Distances   
                            if y[i] != y[j]:
                                distance_array1.append(d) #storing Different Classes Distances
                            else:
                                distance_array2.append(d) #storing Same Classes Distances
                                
                    #Taking the Nearest Distance to Different Class Logic            
                    minimum_index = distance_array1.index(min(distance_array1))
                    diff_class.append(distance_array1[minimum_index])
                    
                    #Taking the Nearest Distance to Same Class Logic
                    minimum_index = distance_array2.index(min(distance_array2))
                    same_class.append(distance_array2[minimum_index])
                    
                    #Reset Arrays
                    distance_array1.clear()
                    distance_array2.clear()
                
                #Logic to find Diff/Same Class
                conformity_scores = [ p/q for p,q in zip(diff_class,same_class) ]   

                test_sample_score = conformity_scores[len(x)-1]
                conformity_scores.sort(reverse=True) #Sort in Descending Order to Calculate rank Pessimistically
                
                for rank in range(len(x)): #Loop to find the rank
                    if conformity_scores[rank]==test_sample_score:
                        break
                        
                rank = len(x)-rank #Logic for Pessimistic Approach to find Rank
                print("Rank of Test Sample "+str(test)+" is: "+str(rank))
                p_value = (rank)/len(x)
                print("P_value: "+str(rank)+"/"+str(len(x))+": ",p_value)
                p_value = p_value*100
                p_value = round(p_value,2)
                print("Percentage: "+str(p_value)+" %")
                print("\n")
                
                #Reset all arrays
                x.clear()
                y.clear()
                diff_class.clear()
                distance_array1.clear()
                distance_array2.clear()
                conformity_scores.clear()
        
    else:
        print("Error:Size of Samples and Labels mismatched")
    
```

```python
CP_DiffBySame(x,y,test)
```
